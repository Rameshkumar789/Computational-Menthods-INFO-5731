```python
import pandas as pd
```

```python
# Read the first 200000 rows
df=pd.read_csv("./Dataset/test-utf8.csv",header=None, sep='\t', skiprows=1, nrows=200000)
```

```python
# Display of first 200,000 rows
df
```

```python
# Assign the column names
df.columns = ['ReferenceID', 'SourceID','ChapterNumber','ParagraphNumber', 'SentenceNumber', 
              'Title','PublishedDate','Authors' ,'TextBeforeRefMention','TextWhereRefMention','TextAfterRefMention']
```

```python
#Display
df
```

```python
# Delete rows which represents tables
df=df.drop(df[df['ReferenceID'].str.match('#tab',na=False)].index)
```

```python
# Delete rows which represents Figures
df=df.drop(df[df['ReferenceID'].str.match('#fig',na=False)].index)
```

```python
# Filter the papers which have citations more than 50   
x=df.groupby(['SourceID']).size().reset_index(name='count')
```

```python
x
```

```python
y=x.loc[x['count'] >= 50, 'SourceID'].values.tolist()
```

```python
# Total Unique papers
len(y)
```

```python
# Filitering
temp=df[df.SourceID.isin(y)]
```

```python
temp
```

```python
# Drop columns
temp = temp.drop(temp.columns[[0, 2, 3,4,6,7,8,10]], axis=1) 
```

```python
#Display
temp
```

```python
#Reset the index
temp = temp.reset_index(drop=True)
```

```python
#Display
temp
```

```python
temp.to_csv("project_50.csv")
```
