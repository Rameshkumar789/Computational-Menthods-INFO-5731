```python
def preprocess_data(doc_set):
    tokenizer = RegexpTokenizer(r'\w+')
    texts = []
    for i in doc_set:
        raw = i.lower()
        tokens = tokenizer.tokenize(raw)
        texts.append(tokens)
    return texts
```

```python
def prepare_corpus(doc_clean):
    dictionary = corpora.Dictionary(doc_clean)
    doc_term_matrix = [dictionary.doc2bow(doc) for doc in doc_clean]
    return dictionary,doc_term_matrix
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.feature_extraction.text import TfidfTransformer
from sklearn import preprocessing
import numpy as np

from sklearn.svm import LinearSVC
from sklearn.neighbors import KNeighborsClassifier

from sklearn.metrics import f1_score
from sklearn.metrics import accuracy_score
from sklearn.metrics import recall_score
from sklearn.metrics import precision_score
```

```python
tfidf_vect = TfidfVectorizer(analyzer='word', token_pattern=r'\w{1,}', max_features=2000)
tfidf_vect.fit(df['Clean_Text'])
```

```python
# Write your code here
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(df['Clean_Text'], df['Sentiment'], test_size=0.2,random_state=50)
```

```python
def transform_train_data(train, test, vect):
    x_train =  vect.transform(train)
    x_test =  tfidf_vect.transform(test)
    return x_train, x_test
```

```python
def transform_test_data(train, test):
  encoder = preprocessing.LabelEncoder()
  train_y = encoder.fit_transform(train)
  valid_y = encoder.fit_transform(test)
  
  return train_y, valid_y
```

```python
x_train, x_test= transform_train_data(X_train, X_test, tfidf_vect)
y_train, y_test = transform_test_data(y_train, y_test)
```

## SVM

```python
clf = LinearSVC()
clf.fit(x_train, y_train)
pred_y = clf.predict(x_test)
print("Accuracy:", accuracy_score(y_test, pred_y))
print("Precision:",precision_score(y_test, pred_y, average='weighted', labels=np.unique(pred_y)))
print("Recall", recall_score(y_test, pred_y, average='weighted'))
print("F1_Score:", f1_score(y_test, pred_y, average='weighted'))
```

## KNN

```python
clf = KNeighborsClassifier()
clf.fit(x_train, y_train)
pred_y = clf.predict(x_test)
print("Accuracy:", accuracy_score(y_test, pred_y))
print("Precision:",precision_score(y_test, pred_y, average='weighted'))
print("Recall", recall_score(y_test, pred_y, average='weighted'))
print("F1_Score:", f1_score(y_test, pred_y, average='weighted'))
```
