```python
import pandas as pd
```

```python
# Read the first 100 rows
df=pd.read_csv("./Dataset/test-utf8.csv",header=None, sep='\t', skiprows=1, nrows=100)
```

```python
# Display of first 100 rows
df
```

```python
# Assign the column names
df.columns = ['ReferenceID', 'SourceID','ChapterNumber','ParagraphNumber', 'SentenceNumber', 
              'Title','PublishedDate','Authors' ,'TextBeforeRefMention','TextWhereRefMention','TextAfterRefMention']
```

```python
#Display
df
```

```python
# Delete rows which represents tables
df=df.drop(df[df['ReferenceID'].str.match('#tab',na=False)].index)
```

```python
# Delete rows which represents Figures
df=df.drop(df[df['ReferenceID'].str.match('#fig',na=False)].index)
```

```python
df
```

```python
# Drop columns
df = df.drop(df.columns[[0, 2, 3,4,6,7,8,10]], axis=1) 
```

```python
#Display
df
```

```python
#Reset the index
df = df.reset_index(drop=True)
```
